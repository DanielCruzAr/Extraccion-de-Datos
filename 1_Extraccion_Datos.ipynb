```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/Detalle precios y productos fabricados 2022.xlsx')
df.head()
```

```python
df.head().T
```

```python
#Filtro por objeto
filtro1 = df[df['NOMBRE_VENDEDOR'] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
filtro1.to_csv("filtro10.csv")
```

```python
#Filtro por filas
filtro2 = df.iloc[3:10,: ]
filtro2
filtro2.to_csv("filtro10.csv")
```

```python
#Filtro por columnas
filtro3 = df.iloc[ : , 1:3]
filtro3
filtro3.to_csv("filtro10.csv")
```

```python
#Filtro por cabecera
filtro5 = df.head()
filtro5
filtro5.to_csv("filtro10.csv")
```

```python
#Filtro por cola
filtro6 = df.tail()
filtro6
filtro6.to_csv("filtro10.csv")
```

```python
#Filtro por comparacion
filtro7 = df[df["PRECIO_UNITARIO"] > 20000]
filtro7
filtro7.to_csv("filtro10.csv")
```

```python
#Filtro Y
filtro8 = df[(df["PRECIO_UNITARIO"] < 2000) & (df["FECHA_DOC"] == "2022-12-23")]
filtro8
filtro8.to_csv("filtro10.csv")
```

```python
#Filtro o
filtro9 = df[(df["PRECIO_UNITARIO"] < 2000) | (df["FECHA_DOC"] == "2022-12-23")]
filtro9
filtro9.to_csv("filtro10.csv")
```

```python
#Filtro not
filtro10 = df[~(df["PRECIO_UNITARIO"] >= 2000) & ~(df["FECHA_DOC"] == "2022-12-23")]
filtro10
filtro10.to_csv("filtro10.csv")
```
