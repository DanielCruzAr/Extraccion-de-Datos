```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Gastos y costos 20-23.xlsx')
```

```python
df.head()
```

```python
# Eliminar las primeras 4 filas
df = df[4:]
df.head()
```

```python
# Hacer que la primera fila sea en encabezado
df.rename(columns = df.iloc[0], inplace=True)
df = df[1:]
df.head()
```

```python
# Revisar valores nulos
df.isnull().sum()
```

##Eliminar valores nulos

```python
# Utilizamos en método bfill
df.fillna(method="bfill", inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# Aplicar ffill en la columna de POLIZA
df.fillna(method="ffill", inplace=True)
df.isnull().sum()
```

```python
# Hacer una gráfica de caja
fig = plt.figure(figsize =(15, 8))
df.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show();
```

##Proceso para eliminar outliers

###Por desviación estándar

```python
# Por std
lim_sup = df.mean() + 3*df.std()
lim_inf = df.mean() - 3*df.std()
print("Limite superior: ", lim_sup)
print("\nLimite inferior: ", lim_inf)
```

```python
#Obtener datos y los outliers se convierten en nulos en el DataFrame
df2 = df[(df<=lim_sup)&(df>=lim_inf)]
df2
```

```python
# Eliminar columnas con NaN
df2 = df2.drop(['FECHA','FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TIPO', 'STATUS', 'POLIZA'], axis=1)
df2
```

```python
# Revisar valores nulos
df2.isnull().sum()
```

```python
# Reemplazar valores atípicos con el promedio
df_clean = df2.copy()
df_clean.fillna(round(df2.mean(),1), inplace=True)
df_clean
```

```python
df_clean.isnull().sum()
```

```python
# Hacer diagrama de caja del df limpio
fig = plt.figure(figsize =(15, 8))
df_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos")
plt.show();
```

```python
cols = df[['FECHA','FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TIPO', 'STATUS', 'POLIZA']]
df_std = pd.concat([cols, df_clean], axis=1)
df_std.head()
```

```python
df_std.to_csv('desviacion_estandar.csv')
```

###Por rango intercuartílico

```python
# Eliminar columnas que no nos sirven
df3 = df.drop(['FECHA','FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TIPO', 'STATUS', 'POLIZA'], axis=1)
df3
```

```python
# Definir cuartiles Q1 y Q3
p25 = df3.quantile(0.25) #Q1
p75 = df3.quantile(0.75) #Q3
# Calcular rango intercuartil
iqr= p75 - p25

# Encontrar límite superior e inferior
lim_sup_iqr = p75 + 1.5 * iqr
lim_inf_iqr= p25 - 1.5 * iqr
print("Limite superior: ")
print(lim_sup_iqr)
print("\nLimite inferior: ")
print(lim_inf_iqr)
```

```python
#Obtener datos limpios
df_iqr = df3[(df3<=lim_sup_iqr)&(df3>=lim_inf_iqr)]
df_iqr
```

```python
df_iqr.isnull().sum()
```

```python
# Reemplazar valores atípicos con promedio
df_iqr = df_iqr.fillna(round(df_iqr.mean(),1))
df_iqr
```

```python
# Diagrama de caja del df
fig = plt.figure(figsize =(15, 8))
df_iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos")
plt.show();
```

```python
cols = df[['FECHA','FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TIPO', 'STATUS', 'POLIZA']]
df_final = pd.concat([cols, df_iqr], axis=1)
df_final
```

```python
# Mandamos a csv
df_final.to_csv('iqr.csv')
```
